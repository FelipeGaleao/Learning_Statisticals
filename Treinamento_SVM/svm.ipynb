```python
import pandas as pd 
import numpy as np 
import sklearn
import matplotlib.pyplot as plt
```

```python
dados = pd.read_excel('Tabela - COR.xlsx')
```

```python
dados
```

```python
dados_x = dados[['COR_2014', 'COR_2015', 'COR_2016', 'COR_2018']].dropna()
dados_y = dados[['COR_2019']].dropna()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dados_x, dados_y[:262], test_size=0.25)
```

```python
x_train.shape
```

```python
from sklearn import svm

modelo = svm.SVR(kernel='linear')

modelo.fit(x_train, y_train.values.ravel())
```

```python
modelo.score(x_train, y_train)
```

```python
y_pred = modelo.predict(x_test)
```

```python
y_pred
```

```python
fig, ax = plt.subplots()

n, bins, patches = ax.hist(y_test, len(y_test), density=True)

ax.plot(y_test, '--')
ax.plot(y_test, '--')

ax.set(xlabel='previsto', ylabel='realizado', title='Modelo SVM')

```
